# Exerceise Regex 

### 1-load the book

```python
with open("miracle_in_the_andes.txt","r", encoding="utf-8") as file:
    book = file.read()
```

### 2 Extract the paragraphs wher "love" was used

```python
import re
pattern = re.compile("[^\n]+love[^\n]+")
findings = re.findall(pattern, book)
findings[:2]
```

### Extract the chapter titles

```python
import re
pattern = re.compile("[a-zA-Z ,]+\n\n")
findings = re.findall(pattern, book)
findings = [item.strip("\n\n") for item in findings]
findings
```

# Method 2

```python
import re
pattern = re.compile("([a-zA-Z]+)\n\n")
findings = re.findall(pattern, book)
findings
```

### Function that finds the occurrence of any word

```python
def find(w):
    pattern = re.compile("([a-zA-Z]+)")
    findings = re.findall(pattern, book)
    d = {}
    for word in findings:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
    try:
        return d[w]
    except:
        return f'The book does not contain the work "{w}"'
```

```python
find("go")
```
