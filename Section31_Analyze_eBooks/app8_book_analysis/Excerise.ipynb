# Exerceise Regex 

### 1-load the book

```python
with open("miracle_in_the_andes.txt","r", encoding="utf-8") as file:
    book = file.read()
```

### 2 Extract the paragraphs wher "love" was used

```python
import re
pattern = re.compile("[^\n]+love[^\n]+")
findings = re.findall(pattern, book)
findings[:2]
```

### Extract the chapter titles

```python
import re
pattern = re.compile("[a-zA-Z ,]+\n\n")
findings = re.findall(pattern, book)
fubdubgs = [item.strip("\n\n") for item in findings]
findings
```
